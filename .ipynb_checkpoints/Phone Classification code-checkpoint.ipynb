```python
categorical_cols = ['blue', 'dual_sim', 'four_g', 'n_cores', 'three_g', 'touch_screen', 'wifi']

for col in categorical_cols:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts())
```

```python
import matplotlib.pyplot as plt

numerical_cols = ['battery_power', 'clock_speed', 'int_memory', 'mobile_wt', 'pc', 
                  'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']

for col in numerical_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    plt.boxplot(df[col], patch_artist=True, boxprops=dict(facecolor='lightgreen'))
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    df.boxplot(column=col, by='price_range', grid=False)
    plt.title(f'{col} by Price Range')
    plt.suptitle('')
    plt.show()
```

```python
for i in range(len(numerical_cols)):
    for j in range(i+1, len(numerical_cols)):
        plt.figure(figsize=(5,4))
        plt.scatter(df[numerical_cols[i]], df[numerical_cols[j]], alpha=0.5)
        plt.xlabel(numerical_cols[i])
        plt.ylabel(numerical_cols[j])
        plt.title(f'{numerical_cols[i]} vs {numerical_cols[j]}')
        plt.show()
```

```python
import numpy as np

corr = df.corr()
plt.figure(figsize=(12,10))
plt.imshow(corr, cmap='coolwarm', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```
