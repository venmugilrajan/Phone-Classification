```python
categorical_cols = ['blue', 'dual_sim', 'four_g', 'n_cores', 'three_g', 'touch_screen', 'wifi']

for col in categorical_cols:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts())
```

```python
df['aspect_ratio'] = df['sc_w'] / df['sc_h']
print(df[['sc_w','sc_h','aspect_ratio']].head())
```

```python
df['total_camera'] = df['fc'] + df['pc']
print(df[['fc','pc','total_camera']].head())
```

```python
df['weight_category'] = pd.cut(df['mobile_wt'], bins=[0, 150, 200, 300], labels=['Light','Medium','Heavy'])
print(df[['mobile_wt','weight_category']].head())
```

```python
df['ram_battery_ratio'] = df['ram'] / df['battery_power']
print(df[['ram','battery_power','ram_battery_ratio']].head())
```

```python
df['connectivity_score'] = df['dual_sim'] + df['four_g'] + df['three_g'] + df['wifi'] + df['blue']
print(df[['dual_sim','four_g','three_g','wifi','blue','connectivity_score']].head())
```

```python
df['price_category'] = df['price_range'].replace({0:'Low',1:'Medium',2:'High',3:'Premium'})
print(df[['price_range','price_category']].head(10))
```

```python
import matplotlib.pyplot as plt

numerical_cols = ['battery_power', 'clock_speed', 'int_memory', 'mobile_wt', 'pc',
                  'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']

for col in numerical_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df['battery_power'], patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Boxplot of Battery Power')
plt.show()
```

```python
plt.figure(figsize=(6,4))
df.boxplot(column='ram', by='price_range', grid=False)
plt.title('RAM by Price Range')
plt.suptitle('')
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(df['battery_power'], df['ram'], alpha=0.5, color='blue')
plt.xlabel('battery_power')
plt.ylabel('ram')
plt.title('battery_power vs ram')
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(df['px_width'], df['px_height'], alpha=0.5, color='red')
plt.xlabel('px_width')
plt.ylabel('px_height')
plt.title('px_width vs px_height')
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(df['clock_speed'], df['battery_power'], alpha=0.5, color='purple')
plt.xlabel('clock_speed')
plt.ylabel('battery_power')
plt.title('clock_speed vs battery_power')
plt.show()
```

```python
df['screen_area'] = df['sc_h'] * df['sc_w']
df['pixel_density'] = np.sqrt(df['px_height']**2 + df['px_width']**2) / df['screen_area']
df['memory_per_core'] = df['int_memory'] / df['n_cores']
df['weight_battery_ratio'] = df['mobile_wt'] / df['battery_power']
df['total_connectivity'] = df['dual_sim'] + df['four_g'] + df['three_g'] + df['wifi'] + df['blue']

plt.figure(figsize=(5,4))
plt.scatter(df['pixel_density'], df['price_range'], color='orange', alpha=0.5)
plt.xlabel('pixel_density')
plt.ylabel('price_range')
plt.title('Pixel Density vs Price Range')
plt.show()
```

```python
df=pd.read_csv('/content/Phone-Classification.csv')

numerical_cols = df.select_dtypes(include=np.number).columns
corr = df[numerical_cols].corr()

plt.figure(figsize=(10,8))
plt.imshow(corr, cmap='coolwarm', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Correlation Heatmap (Numerical Columns Only)")
plt.show()
```
